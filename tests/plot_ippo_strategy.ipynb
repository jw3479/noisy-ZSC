```python
def plot_traj(df):
    dat = df.iloc[::100, :]
    plt.scatter(dat.index, dat.true_max, label = "true max", alpha = 0.2)
    plt.scatter(dat.index, dat.E1_max, label = "E1 max", alpha = 0.2)
    plt.scatter(dat.index, dat.E2_max, label = "E2 max", alpha = 0.2)
    plt.scatter(dat.index, dat.E1_val, label = "E1 val", alpha = 0.2)
    plt.scatter(dat.index, dat.E2_val, label = "E2 val", alpha = 0.2)
    plt.scatter(dat.index, dat.reward_list, label = "reward", alpha = 0.2)
    #plt.plot(df.E1_max, label = "E1_max")
    #plt.plot(df.E2_max, label = "E2_max")
    plt.legend()
    plt.show()
    
```
